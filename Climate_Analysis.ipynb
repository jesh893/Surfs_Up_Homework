```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
import datetime as dt
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
 # Create our session (link) from Python to the DB
session = Session(engine)
```

```python
first_entry = session.query(Measurement).first()
first_entry
```

```python
# Print the first entry in Dictionary format.
first_entry.__dict__
```

```python
# Determine the latest entry Date. 
latest_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
latest_date
```

```python
# Print precipitation results for the exactly 12 months from 08-23-2017. 
yearly_precipitation = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= "2016-08-23").group_by(Measurement.date).all()
yearly_precipitation
```

```python
# Put the Precipitation results into neat Data Frame. 
precipitation_df = pd.DataFrame(yearly_precipitation)
precipitation_df.head()
```

```python
# The Data Frame with Date Index. 
precipitation_index = precipitation_df.set_index("date")
precipitation_index.head()
```

```python
# Graph the results using .plot method. Set the Graph Title, Save the figure in the Homework Folder and show the graph. 

precipitation_index.plot()

plt.title("Total Precipitation")

plt.savefig("Total Precipitation Graph")

plt.show()
```

```python
# summary statistics for the precipitation data
summary_stats = precipitation_index.describe()
summary_stats
```

```python
#Station Analysis
```

```python
station_first = session.query(Station).first()
station_first
```

```python
station_first.__dict__
```

```python
# Total Number of Station
total_stations = session.query(Station.station).count()
total_stations
```

```python
# Determine the most active stations and create a list. 

active_station = session.query(Measurement.station,func.count(Measurement.station)).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
        
active_station
```

```python
# Single most active Station
most_active_station = active_station[0][0]
most_active_station
```

```python
# Temperature Observation Data or tobs
temperature_observation_data = session.query(Measurement.tobs).filter(Measurement.date >= "2016-08-23", Measurement.station == most_active_station).all()
temperature_observation_data
```

```python
tobs_df = pd.DataFrame(temperature_observation_data)
tobs_df.head()
```

```python
# Plot the graph and save the Image. 

plt.hist(tobs_df["tobs"], bins=12)

plt.savefig("Tobs")

plt.show()
```
